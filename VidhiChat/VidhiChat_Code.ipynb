```python
from langchain_community.document_loaders import TextLoader
from langchain.embeddings import HuggingFaceEmbeddings

loader = DirectoryLoader('/content/untitled/desired_text_files_2',glob="./*.txt", loader_cls=TextLoader)

from langchain_community.vectorstores import FAISS
from langchain_text_splitters import CharacterTextSplitter

documents = loader.load()
text_splitter = CharacterTextSplitter(chunk_size=1000, chunk_overlap=0)
texts = text_splitter.split_documents(documents)
embeddings = HuggingFaceEmbeddings(model_name="BAAI/bge-base-en-v1.5")
persist_directory = 'db'

## Here is the nmew embeddings being used
embedding = embeddings #instructor_embeddings

db = Chroma.from_documents(documents=texts,
                                 embedding=embedding,
                                 persist_directory=persist_directory)
```

```python
from langchain.schema.runnable import RunnablePassthrough
from langchain.schema.output_parser import StrOutputParser
from langchain.prompts import ChatPromptTemplate
```

```python
template = """
 <|system|>
    You are an AI assistant that follows instruction extremely well.
    Strictly provide citation for the documents that you used as reference.
    Please be truthful and give direct answers. Do not use your own knowledge use from the knowledgebase.
    Keep the answer *length below strictly below 200 words*, and be sure to add the reference and file name after the answer.
    If no answer is available from the context provided please strictly say "The answer cannot be found in the context provided".
 <|user|>
 Context: {context}
 Query: {query}
 </s>
 <|assistant|>
"""
```

```python
!pip install langchainhub --quiet
from langchain import hub
prompt = hub.pull("rlm/rag-prompt")
```

```python
prompt = ChatPromptTemplate.from_template(template)
```

```python
def format_docs(docs):
    return "\n\n".join(documents.page_content for documents in docs)
```

```python
class ChatBot():
  rag_chain = (
    {"context": retriever,  "query": RunnablePassthrough()}
    | prompt
    | llm
    | StrOutputParser()
  )
```

```python
import gradio as gr
bot = ChatBot()

# Function for generating LLM response
def generate_response(input):
    result = bot.rag_chain.invoke(input)
    # Split the result by the unique string and return the desired portion
    try:
        result = result.split('<|assistant|>')[-1].strip()
    except IndexError:
        result = "Error: Unique string not found in the response."
    return result

# Gradio interface
with gr.Blocks() as demo:
    gr.Markdown("# Bot for Indian Legal System")

    # Chatbot UI elements
    chatbot = gr.Chatbot()
    user_input = gr.Textbox(show_label=False, placeholder="Enter your query and press enter")

    # Function to handle user input and update the chat
    def handle_input(input_text, chat_history):
        if input_text:
            # Generate response using RAG
            result = generate_response(input_text)
            # Append user and assistant messages to chat history
            chat_history.append(("user", input_text))
            chat_history.append(("assistant", result))
        # Return updated chat history
        return chat_history

    # Set up the interaction
    user_input.submit(handle_input, [user_input, chatbot], chatbot)

demo.launch()
```
